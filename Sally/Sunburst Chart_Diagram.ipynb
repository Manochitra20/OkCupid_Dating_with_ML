```python
import pandas as pd
import numpy as np

pd.set_option("max_columns", 30)

import plotly.express as px
import plotly.graph_objects as go
```

```python
OKCupid_locations = pd.read_csv("OKcupid_cleaned_up.csv")
OKCupid_locations.head()
```

```python
OKCupid_locations.rename( columns={'Unnamed: 0': 'Okcupid'}, inplace=True )
OKCupid_locations.head()
```

# Plotly Express

```python
OKCupid_dist = OKCupid_locations.groupby(by=["location", "body_type", "job"]).count()
OKCupid_dist ['Okcupid'] = "Okcupid"
OKCupid_dist = OKCupid_dist .reset_index()
OKCupid_dist.head()
```

```python
fig = px.sunburst(OKCupid_dist,
                  path=["Okcupid", "location", "body_type", "job"],
                  values='Count',
                  title="Okcupid App[location, body_type, job]",
                  width=750, height=750)
fig.show()
```
