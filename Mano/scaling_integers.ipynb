```python
# Initial imports
import pandas as pd
from sklearn.cluster import KMeans
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
# Loading data
file_path = Path("cluster_clean.csv")
df = pd.read_csv(file_path)
df.head()
```

```python
# Scale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[['age', 'height', 'income']])
```

```python
# A list of the columns from the original DataFrame
df.columns
```

```python
# Create a DataFrame with the transformed data
new_df=pd.DataFrame((scaled_data), columns={'age_scaled':'age',
                                           'height_scaled': 'height',
                                           'income_scaled': 'income'})
new_df.head()
```

```python
#joining two dataframes
join_df=df.join(new_df)
join_df.head()
```

```python
join_df1=join_df.drop(columns=['age', 'height', 'income'], axis=1)
join_df1.head()
```

```python
# Saving cleaned data
file_path = Path("scaled_profile.csv")
join_df1.to_csv(file_path, index=False)
```
