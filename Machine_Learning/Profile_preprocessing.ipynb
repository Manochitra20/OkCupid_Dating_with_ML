```python
# Transform Previous Customer column
def changeStatus(status):
    if status == "m":
        return 1
    else:
        return 0

# Along with replace() and map(), this is another way to encode the gender column into numbers.
cluster_clean["sex"] = cluster_clean["sex"].apply(changeStatus)
cluster_clean.head(10)
```

```python
from sklearn.preprocessing import OneHotEncoder

# Instantiate OneHotEncoder
bodytype_ohe = OneHotEncoder()
diet_ohe = OneHotEncoder()
drinks_ohe = OneHotEncoder()
drugs_ohe = OneHotEncoder()
education_ohe = OneHotEncoder() 
ethnicity_ohe = OneHotEncoder()
job_ohe = OneHotEncoder()
orientation_ohe = OneHotEncoder()
religion_ohe = OneHotEncoder()
smokes_ohe = OneHotEncoder()
status_ohe = OneHotEncoder()
sign_ohe = OneHotEncoder()


# Fit the OneHotEncoder and transform
bodytype = pd.DataFrame(cluster_clean['body_type'])
bodytype_encoded = bodytype_ohe.fit_transform(bodytype)
display(bodytype_encoded)

diet = pd.DataFrame(cluster_clean['diet'])
diet_encoded = diet_ohe.fit_transform(diet)
display(diet_encoded)

drinks = pd.DataFrame(cluster_clean['drinks'])
drinks_encoded = drinks_ohe.fit_transform(drinks)
display(drinks_encoded)

drugs = pd.DataFrame(cluster_clean['drugs'])
drugs_encoded = drugs_ohe.fit_transform(drugs)
display(drugs_encoded)

education = pd.DataFrame(cluster_clean['education'])
education_encoded = education_ohe.fit_transform(education)
display(education_encoded)

ethnicity = pd.DataFrame(cluster_clean['ethnicity'])
ethnicity_encoded = ethnicity_ohe.fit_transform(ethnicity)
display(ethnicity_encoded)

job = pd.DataFrame(cluster_clean['job'])
job_encoded = job_ohe.fit_transform(job)
display(job_encoded) 

orientation = pd.DataFrame(cluster_clean['orientation'])
orientation_encoded = orientation_ohe.fit_transform(orientation)
display(orientation_encoded)

religion = pd.DataFrame(cluster_clean['religion'])
religion_encoded = religion_ohe.fit_transform(religion)
display(religion_encoded) 

smokes = pd.DataFrame(cluster_clean['smokes'])
smokes_encoded = smokes_ohe.fit_transform(smokes)
display(smokes_encoded)

status = pd.DataFrame(cluster_clean['status'])
status_encoded = status_ohe.fit_transform(status)
display(status_encoded)

sign = pd.DataFrame(cluster_clean['sign_cleaned'])
sign_encoded = sign_ohe.fit_transform(sign)
display(sign_encoded)
```

```python
# Put into a dataframe to get column names
#bodytype
bodytype_df = pd.DataFrame(bodytype_encoded.toarray().astype(int), columns=bodytype_ohe.categories_[0], dtype=int)
# drop first column for redundancy
#bodytype_df = bodytype_df.drop(bodytype_df.columns[0], axis=1)
display(bodytype_df.head())

#diet
diet_df = pd.DataFrame(diet_encoded.toarray().astype(int), columns=diet_ohe.categories_[0], dtype=int)
#diet_df = diet_df.drop(diet_df.columns[0], axis=1)
display(diet_df.head())

#drinks
drinks_df = pd.DataFrame(drinks_encoded.toarray().astype(int), columns=drinks_ohe.categories_[0], dtype=int)
#drinks_df = drinks_df.drop(drinks_df.columns[0], axis=1)
display(drinks_df.head())

#drugs
drugs_df = pd.DataFrame(drugs_encoded.toarray().astype(int), columns=drugs_ohe.categories_[0], dtype=int)
#drugs_encoded = drugs_df.drop(drugs_df.columns[0], axis=1)
display(drugs_df.head())

#education
education_df = pd.DataFrame(education_encoded.toarray().astype(int), columns=education_ohe.categories_[0], dtype=int)
#education_df = education_df.drop(education_df.columns[0], axis=1)
display(education_df.head())

#ethnicity
ethnicity_df = pd.DataFrame(ethnicity_encoded.toarray().astype(int), columns=ethnicity_ohe.categories_[0], dtype=int)
#ethnicity_df = ethnicity_df.drop(ethnicity_df.columns[0], axis=1)
display(ethnicity_df.head())

# Jobs
job_df = pd.DataFrame(job_encoded.toarray().astype(int), columns=job_ohe.categories_[0], dtype=int)
#job_df = job_df.drop(job_df.columns[0], axis=1)
display(job_df.head())

#orientation
orientation_df = pd.DataFrame(orientation_encoded.toarray().astype(int), columns=orientation_ohe.categories_[0], dtype=int)
#orientation_df = orientation_df.drop(orientation_df.columns[0], axis=1)
display(orientation_df.head())

# Status
status_df = pd.DataFrame(status_encoded.toarray().astype(int), columns=status_ohe.categories_[0], dtype=int)
#status_df = status_df.drop(status_df.columns[0], axis=1)
display(status_df.head())

#religion
religion_df = pd.DataFrame(religion_encoded.toarray().astype(int), columns=religion_ohe.categories_[0], dtype=int)
#religion_df = religion_df.drop(religion_df.columns[0], axis=1)
display(religion_df.head())

#smokes
smokes_df = pd.DataFrame(smokes_encoded.toarray().astype(int), columns=smokes_ohe.categories_[0], dtype=int)
#smokes_df = smokes_df.drop(smokes_df.columns[0], axis=1)
display(smokes_df.head())

#sign
sign_df = pd.DataFrame(sign_encoded.toarray().astype(int), columns=sign_ohe.categories_[0], dtype=int)
#sign_df = sign_df.drop(sign_df.columns[0], axis=1)
display(sign_df.head())

```
